#Problem 1(a).  Word length.

Write a function called ```word_length_list()``` which takes a string and returns a list with the length of each word in the string.  Your function should ignore punctuation.  For example, ```word_length_list("The panda eats, shoots, and leaves!")``` should return ```[3,5,4,6,3,6]```.  

Call or create other functions as necessary to organize your work.

````python
import string
def  word_length_list(str):
    lst = []
    g = str.split(" ")
    
    for word in g:
        count = 0
        for character in word:
            if character not in string.punctuation:
                count = count + 1
        lst.append(count)
    return(lst)
                
print(word_length_list("The panda eats, shoots, and leaves!"))
````

#Problem 1(c). _Pride and Prejudice_.

Use Python to count the number of words and mean length of words in each sentence of Pride and Prejudice (https://www.gutenberg.org/cache/epub/1342/pg1342.txt).  Save your result as a ```.csv``` file.

* Create new functions as necessary to organize your work.
* Include comments to explain the purpose and arguments of each function you create.

````python
import re
import csv
import numpy

file = open("Pride&Prejudice.txt","r")
doclist = [line for line in file]
#joins every new line
docstr = ''.join(doclist)
#split by regex with period,exclamation and question marks
sentences = re.split(r'[.?!]', docstr)


#for s in sentences:
    #print(s)
    #print(s.split(' '))

csv_wordcount = []
csv_avg_word = []
for i in sentences:
    
    #splits each word in the list to it's own index
    words = i.split(' ')
    csv_wordcount.append(len(words))
    
    # a sum counter 
    sum_T = 0
    for j in i:
        sum_T +=len(j)
    # NO NEED TO PRINT this print(sum_T)
    csv_avg_word.append(sum_T/len(words))
    

    
dualList = zip(csv_wordcount, csv_avg_word)
with open("1c.csv", "w" ,newline="") as out:
    writer = csv.writer(out)
    for index, (i,j) in enumerate(dualList):
        writer.writerow([i,j])
    
````
