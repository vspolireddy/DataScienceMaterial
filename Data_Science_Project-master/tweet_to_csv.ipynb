```python
import pandas as pd
import json
import re
import csv
 
# A regex converted that filters out entries that are not in city, State format
# we need this later on to extract long and lat for each location
def regex_getter(entry):
    
    txt = entry

    re1='((?:[a-z][a-z]+))'	# Word 1
    re2='(,)'	# Any Single Character 1
    re3='(\\s+)'	# White Space 1
    re4='((?:(?:AL)|(?:AK)|(?:AS)|(?:AZ)|(?:AR)|(?:CA)|(?:CO)|(?:CT)|(?:DE)|(?:DC)|(?:FM)|(?:FL)|(?:GA)|(?:GU)|(?:HI)|(?:ID)|(?:IL)|(?:IN)|(?:IA)|(?:KS)|(?:KY)|(?:LA)|(?:ME)|(?:MH)|(?:MD)|(?:MA)|(?:MI)|(?:MN)|(?:MS)|(?:MO)|(?:MT)|(?:NE)|(?:NV)|(?:NH)|(?:NJ)|(?:NM)|(?:NY)|(?:NC)|(?:ND)|(?:MP)|(?:OH)|(?:OK)|(?:OR)|(?:PW)|(?:PA)|(?:PR)|(?:RI)|(?:SC)|(?:SD)|(?:TN)|(?:TX)|(?:UT)|(?:VT)|(?:VI)|(?:VA)|(?:WA)|(?:WV)|(?:WI)|(?:WY)))(?![a-z])'	# US State 1

    rg = re.compile(re1+re2+re3+re4,re.IGNORECASE|re.DOTALL)
    m = rg.search(txt)
    if m:
        word1=m.group(1)
        c1=m.group(2)
        ws1=m.group(3)
        usstate1=m.group(4)
        return str((word1 +  c1 +ws1 + usstate1))



# import json package to load json file
#with open('bernie_raw_tweets.json',encoding="utf8") as jsonFile:
with open('donald_raw_tweets.json',encoding="utf8") as jsonFile:
    polls_json = json.loads('[{}]'.format(jsonFile.read()))

# Get the coulmns user and location from the data frame and extract 
tweets_list = [(dct['user']['location'], dct["text"]) for dct in polls_json]
colNames = ("location", "text")
tweets_df = pd.DataFrame(tweets_list, columns=colNames)


tweets_df=tweets_df.dropna()

tweets_df.head()

df = pd.DataFrame([])

#gets and prints the split of city and state
for i in tweets_df.ix[:,0]:
    if regex_getter(i) != None:
        city,state = regex_getter(i).split(',')
        #print(city,state)

        
#open a csv and write cities and states 
with open("city_states.csv", "w", newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['City', 'State','Color'])
    for i in tweets_df.ix[:,0]:
        if regex_getter(i) != None:
            data = regex_getter(i).split(',')
            #blue or red for GOP or Democratic
            data.append('red')
            #strip white spaces from column 2
            data[1] = data[1].strip()
            print(data)
            writer.writerow(data)
            
            

            


```
