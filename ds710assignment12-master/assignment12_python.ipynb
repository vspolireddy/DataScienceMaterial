#Problem 1(a).  A collectible card class.

You are tasked with organizing information about a Magic: The Gathering card collection.  You are particularly interested in which cards are most rare, and therefore most valuable.

The MTG JSON website (http://mtgjson.com/) has information about different sets of Magic: The Gathering cards in JSON format.  Using the [Example Card](http://mtgjson.com/#exampleCard) information, create a MagicCard class to store relevant information about Magic cards in Python.

Consider:

* How should someone initialize an instance of your class?
* What functions should your class have?

```python
class MagicCard(object):
    
    def __init__(self, id, name, cmc, colors, type, rarity, number):
        self.id = id
        self.name = name
        self.cost = cmc
        self.colors = colors
        self.type = type
        self.rarity = rarity
        self.number = number
    
    
# Access methods, getters    
    def get_id(self):
        """ return card id"""
        return self.id
      
    def get_name(self):
        """ return card name"""
        return self.name
    
    def get_cost(self):
        """ return converted mana cost"""
        return self.cost
    
    def get_colors(self):
        """ return card colors"""
        return self.colors
    
    def get_type(self):
        """ return card type"""
        return self.type
    
    def get_rarity(self):
        """ return card rarity"""
        return self.rarity
    
    def get_number(self):
        """ return card number"""
        return self.number
```

#Problem 1(b). Class for a set of Magic cards.

Download the data for the Dragons of Tarkir set from the [Individual Sets](http://mtgjson.com/#individualSets) section of the MTG JSON page.  Create a MagicCardSet class to store relevant information about a set of Magic cards, and create an instance of the class which contains all of the information about the Dragons of Tarkir set.

```python
class magicCards(object):
    
    def __init__(self, name="", cardType="", colors="", rarity=""):
        self.name = name
        self.type = cardType
        self.colors = colors
        self.rarity = rarity
    
    def getColors(self):
        """ returns card color"""
        return self.colors
    
    def getRarity(self):
        """ returns card's rarity """
        return self.rarity
    
    def getName(self):
        """ returns card name """
        return self.name
    
    def getCardType(self):
        """ returns card card """
        return self.type
    
    
```

```python
import json
import pandas as panda


# import json package to load json file
with open('DTK.json',encoding="utf8") as jsonFile:
    mtg_json = json.loads(jsonFile.read())

```

```python
# list comprehension to itereate through the mtg_json object, each iteration is a card   

cards_list = [magicCards(card["name"], card["type"], card.get("colors"), card["rarity"]) for card in mtg_json['cards']]
```

```python
# col names be used for data Frame
colNames = ("Name", "CardType", "Color", "Rarity")
dict_list = []

# for each card we extract the fields needed 
for card in cards_list:
    dict_list.append(dict(zip(colNames , [card.getName(), card.getCardType(), card.getColors(), card.getRarity()])))
magic_dataFrame = panda.DataFrame(dict_list)
magic_dataFrame.head()
```
